## Task: Generate clusters with make_blobs function 

Use the following settings:
   * n_samples=150,
   * n_features=2,
   * centers=3,
   * cluster_std=0.6,
   * random_state=0

```python
# import libraries
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

# generate clusters 
X, y = make_blobs(n_samples=150,
                  n_features=2,
                  centers=3,
                  cluster_std=0.6,
                  random_state=0)

```

* plot data points

```python

# plot data
plt.scatter(X[:,0], X[:,1])
plt.grid()
plt.show()
```

* define function plot_clusters

```python
def plot_clusters(X,y_res, plt_cluster_centers = False):
    X_centroids = []
    Y_centroids = []

    for cluster in set(y_res):
        x = X[y_res == cluster,0]
        y = X[y_res == cluster,1]
        X_centroids.append(np.mean(x))
        Y_centroids.append(np.mean(y))

        plt.scatter(x,
                    y,
                    s=50,
                    marker='s',
                    label=f'cluster {cluster}')

    if plt_cluster_centers:
        plt.scatter(X_centroids,
                    Y_centroids,
                    marker='*',
                    c='red',
                    s=250,
                    label='centroids')
    plt.legend()
    plt.grid()
    plt.show()


```

## Objective: 
Try to find optimal settings of `eps` and `min_samples` params of DBSCAN. You can iterate over multiple parameters combinations and plot the clustering results.

* define params space: two lists with `eps` and `min_samples` respectively

```python
eps=[i/10 for i in range(1,8)]
eps
```

```python
min_samples=[i for i in range(1,6)]
min_samples
```

* combinations (hint: use 'product' function from 'itertools')

```python
from itertools import product
for i in list(product(eps, min_samples)):
    db = DBSCAN(eps=i[0],
            min_samples=i[1],
            metric='euclidean')
# fit and predict
    y_db = db.fit_predict(X)
# Plot DBSCAN clusters
    plot_clusters(X,y_db)
    print(i)
    
```
